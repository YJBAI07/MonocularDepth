```python

import os, re, pathlib
out = pathlib.Path("/kaggle/working/nyuv2_stylized_adain_8")
for p in out.glob("*.png"):
    # 把文件名里第一次出现的 _stylized_ 及其后面全部去掉
    new = re.sub(r"_stylized_.*$", "", p.stem) + p.suffix
    if new != p.name:
        p.rename(p.with_name(new))
print("Renamed.")
```

```python
import shutil

folder = "/kaggle/working/nyuv2_stylized_adain_8"
zip_base = "/kaggle/working/nyuv2_stylized_adain_8"

shutil.make_archive(zip_base, "zip", root_dir=folder)

print(f"打包完成: {zip_base}.zip")
```
